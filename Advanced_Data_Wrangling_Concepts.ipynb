### **Another way to use apply() involves the use of lambda() function.**

* A lambda() function is a small anonymous function.

Lambda functions are useful when you need to perform a quick operation on a piece of data, like ‚ûï adding or ‚úñ multiplying, but you don't want to define a whole new function just for that one task. Think of it like calling in a ninja ü•∑ to perform a quick strike and disappear without a trace!

```python
# lambda function - Row wise operation #When 2 or more columns are involved
def revised_profit(revenue, budget):
  if revenue > 0:
    new_profit = revenue - budget
  else:
    new_profit = np.nan

  return new_profit
```

```python
imdb_df[['revenue','budget']].apply(revised_profit)
```

```python
imdb_df['new_profit_1'] = imdb_df.apply(lambda x: revised_profit(x['revenue'], x['budget']),axis = 1)
imdb_df[['revenue','budget',"new_profit_1"]]
```

This code defines a function called revised_profit that takes two arguments, revenue and budget, and returns a new profit value if the revenue is greater than 0, or NaN if the revenue is less than or equal to 0.

The apply() function is then used on the imdb_df DataFrame to apply the revised_profit function to each row of the DataFrame. The lambda function takes each row of the DataFrame as input and passes the values of the revenue and budget columns to the revised_profit function. The result of the function is then assigned to a new column in the DataFrame called new_profit.

## **Now let's do some joins**




üîó In the context of Pandas, joins refer to operations that combine two or more dataframes based on common columns or indices. The merge() function in Pandas can be used to perform these joins.

üî¢ Pandas supports four types of joins:

üîç Inner join: Returns only the matching rows from both dataframes based on a common column or index.

üåÄ Outer join: Returns all the rows from both dataframes and fills the missing values with NaN if a row is unmatched.

üëà Left join: Returns all the rows from the left dataframe and matching rows from the right dataframe. Missing values are filled with NaN if a row is unmatched.

üëâ Right join: Returns all the rows from the right dataframe and matching rows from the left dataframe. Missing values are filled with NaN if a row is unmatched.

ü§î The type of join to use depends on the analysis being performed and the specific requirements of the task at hand.

### <B> Using **Inner Join** to merge the two dataframes

An inner join ü§ù refers to the process of combining two dataframes based on the values of a common column üîó, and only keeping the rows where there is a match üéØ between the values in that column in both dataframes.

PostgreSQL-Join-Inner-Join.png

Here is how? üëá

```python
merged_IJ_df = pd.merge(english_movies,long_movies,how='inner',left_on='imdb_id',right_on='imdb_id')
merged_IJ_df.head()
```

```python
merged_IJ_df['runtime'].min()
```

This code is merging two pandas dataframes, english_movies and long_movies, based on the shared column imdb_id. The resulting merged dataframe will contain only the rows that have matching imdb_id values in both input dataframes.

The how='inner' parameter specifies that an inner join should be used, meaning only the rows that have matching imdb_id values in both dataframes will be included in the merged dataframe.

The left_on='imdb_id' and right_on='imdb_id' parameters specify that the join should be based on the imdb_id column in both dataframes, as this is the shared column that will be used to match rows between the two dataframes.

The resulting dataframe merged_IJ_df will have the columns from both dataframes but only the rows that had matching imdb_id values.

### <b> Using left join to merge the two dataframes

ü§îüîç When it comes to code, performing a left join in Pandas requires using the .merge() method and specifying the left and right dataframes, the common key(s) to merge on, and the type of join to perform (in this case, how='left').

ü§ùüíª This process compares the common key(s) in the left and right dataframes and returns a new dataframe that includes all the rows from the left dataframe and the matching rows from the right dataframe.

üí°üìä However, for any non-matching rows, the columns corresponding to the right dataframe will have null values in the resulting dataframe.

lrf.png

Let's perform left join üëá

```python
# Merging the two dataframes using left join
merged_LJ_df = pd.merge(english_movies, long_movies, how='left',left_on='imdb_id',right_on='imdb_id')
```

This code performs a left join on two dataframes english_movies and long_movies using the **.merge()** method in Pandas.

The resulting merged dataframe is assigned to **merged_LJ_df**.

The **how='left'** parameter specifies that we want to perform a left join, which will include all rows from the left dataframe **(english_movies)** and only matching rows from the right dataframe **(long_movies).**

The **left_on** and **right_on** parameters specify the column(s) to use as keys for the join operation. In this case, the **imdb_id** column is used as the common key for both dataframes.

By using the left join, we retain all rows from **english_movies** and only include matching rows from **long_movies**, while replacing non-matching rows with null values in the columns corresponding to **long_movies**.

```python
merged_LJ_df.isnull()
```

### <b> Using right join to merge the two dataframes

üßêüîó A right join, also known as a right outer join, is a type of join operation that is used to combine two dataframes based on a common key, but it includes all rows from the right dataframe and only the matching rows from the left dataframe.

In other words, a right join returns all the rows from the right dataframe and fills in any matching rows from the left dataframe, and replaces any non-matching rows from the left dataframe with null values.

To perform a right join in pandas, you can use the .merge() method and specify the right and left dataframes, the common key(s) to merge on, and the type of join to perform (in this case, how='right'). ü§ùüíª

ryt.png

```python
merged_RJ_df= pd.merge(english_movies,long_movies,how='right',left_on='imdb_id',right_on='imdb_id')
merged_RJ_df
```

This code performs a right join on two dataframes **english_movies** and **long_movies** using the **.merge()** method in Pandas.

The resulting merged dataframe is assigned to **merged_RJ_df**.

The **how='right'** parameter specifies that we want to perform a right join, which will include all rows from the right dataframe **(long_movies)** and only matching rows from the left dataframe **(english_movies)**.

The **left_on** and **right_on** parameters specify the column(s) to use as keys for the join operation. In this case, the **imdb_id** column is used as the common key for both dataframes.

By using the right join, we retain all rows from **long_movies** and only include matching rows from **english_movies**, while replacing non-matching rows from english_movies with null values.

```python
merged_RJ_df
```

### <b> Using outer join to merge the two dataframes

fyll.png

üîçü§î An outer join, also known as a full outer join, is a type of join operation used to combine two dataframes based on a common key(s). In an outer join, all rows from both dataframes are included in the resulting dataframe, regardless of whether there is a match or not.

üíª In Pandas, you can perform an outer join using the .merge() method and specifying the left and right dataframes, the common key(s) to merge on, and the type of join to perform (in this case, how='outer').

üßê The resulting dataframe will have all the rows from both dataframes, with null or NaN values in the columns corresponding to any non-matching rows.

```python
# Outer Join
merged_OJ_df = pd.merge(english_movies[['imdb_id','runtime']],long_movies, on='imdb_id', how='outer')
```

This code performs an outer join on two dataframes **english_movies** and **long_movies** using the **.merge()** method in Pandas.

The  **[ ['imdb_id','runtime'] ]** syntax is used to subset the columns in the **english_movies** dataframe to only include the **imdb_id** and runtime columns. This subset dataframe is used as the left dataframe in the join operation.

The **on='imdb_id'** parameter specifies that we want to perform the join on the **imdb_id** column, which is the common key between the two dataframes.

The **how='outer'** parameter specifies that we want to perform an outer join, which will include all rows from both dataframes, regardless of whether there is a match or not.

The resulting merged dataframe is assigned to **merged_OJ_df**.

The resulting dataframe will have all the rows from both dataframes, with **null** or **NaN** values in the columns corresponding to any non-matching rows.

```python
merged_OJ_df
```

Have you ever played a game of Tetris? üïπÔ∏èüëæ Well, groupby() is like playing Tetris with your data! Imagine that each block in Tetris is a row in your dataset, and you want to group those rows together based on a specific characteristic.

üîç For example, let's say you have a dataset of movies, and you want to group them by genre. You can use the groupby() operation to stack all the movies with the same genre on top of each other, like Tetris blocks!

Here's how it works:

First, you select the column that you want to group by, such as the genre column.
Then, you call the groupby() function on your dataframe and pass in the name of the column you want to group by.
Finally, you can perform aggregate functions on the grouped data, such as calculating the average rating for each genre

üêºüíª Group by operations in Pandas refer to the process of grouping rows in a DataFrame based on the values in one or more columns, and then applying some aggregate functions to the grouped data.

üë®‚Äçüíºüìä The groupby() function in Pandas is like a manager that organizes your data into groups based on common characteristics. The output of the groupby() function is a DataFrameGroupBy object, which can be used to apply various aggregate functions such as sum(), mean(), count(), max(), min(), and many more.
