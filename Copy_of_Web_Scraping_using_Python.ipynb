##<b>Understanding HTML Structure

In order to effectively scrape data from a website, it's important to understand its HTML structure. Here are some key concepts to keep in mind:

###**1. HTML tags and attributes**:



 HTML is the language used to create websites, and it consists of a series of tags and attributes that define the structure and content of a webpage. Tags are used to create elements such as headings, paragraphs, and images, while attributes provide additional information about these elements, such as their size, color, or position on the page.

* ```<h1>``` - This is a tag used to create a heading on a webpage. The number after the 'h' indicates the level of the heading (e.g. ```<h1>``` is the highest level heading).
* ```<p>``` - This is a tag used to create a paragraph on a webpage.
* ```<img src="image.jpg" alt="description">``` - This is a tag used to insert an image on a webpage. The 'src' attribute specifies the location of the image file, while the 'alt' attribute provides a description of the image for users who cannot see it.

* ```<a href="https://example.com">link</a>``` - This is a tag used to create a hyperlink on a webpage. The 'href' attribute specifies the URL that the link should point to.

* ```<div>``` - This is a tag used to create a container on a webpage. It can be used to group related elements together and apply styles or formatting to them.
* ```<table>``` - This is a tag used to create a table on a webpage. It can contain ```<tr>``` (table row) and * ```<td>``` (table data) tags to create rows and cells within the table.
* ```<form action="submit.php" method="post">``` - This is a tag used to create a form on a webpage. The 'action' attribute specifies the URL that the form should be submitted to, while the 'method' attribute specifies the HTTP method to use (e.g. 'post' or 'get').
* ```<input type="text" name="username">``` - This is a tag used to create an input field on a webpage. The 'type' attribute specifies the type of input (e.g. 'text' or 'password'), while the 'name' attribute specifies the name of the input field that will be submitted when the form is submitted.

These are just a few examples of the many HTML tags and attributes that can be used to create and format content on a webpage.





###**2. Identifying elements to scrape**:



When scraping a website, it's important to identify the specific elements you want to extract. This can include text, images, links, or any other content that is relevant to your scraping task. You can use tools like BeautifulSoup or Scrapy to locate these elements in the website's HTML code.

* **Text**: To scrape text from a webpage, you can identify the HTML tags that contain the text you want to extract. For example, if you want to scrape the product names from an e-commerce website, you might look for ```<h2>``` tags that contain the product names. You can use BeautifulSoup or Scrapy to locate these tags and extract the text within them.

* **Images**: To scrape images from a webpage, you can identify the HTML tags that contain the image URLs. For example, if you want to scrape the product images from an e-commerce website, you might look for ```<img>``` tags that contain the image URLs. You can use BeautifulSoup or Scrapy to locate these tags and extract the 'src' attribute, which contains the image URL.

* **Links**: To scrape links from a webpage, you can identify the HTML tags that contain the link URLs. For example, if you want to scrape the links to product pages on an e-commerce website, you might look for ```<a>``` tags that contain the product URLs. You can use BeautifulSoup or Scrapy to locate these tags and extract the 'href' attribute, which contains the link URL.

* **Tables**: To scrape data from a table on a webpage, you can identify the HTML tags that contain the table rows and cells. For example, if you want to scrape the prices and ratings of products from an e-commerce website, you might look for ```<table>``` tags that contain ```<tr>``` (table row) and ```<td>``` (table data) tags. You can use BeautifulSoup or Scrapy to locate these tags and extract the text within them.

By identifying the specific elements you want to scrape and using tools like BeautifulSoup or Scrapy to locate them in the website's HTML code, you can more effectively extract the data you need for your scraping task.

Once installed, you can start using Beautiful Soup to parse and extract data from HTML or XML documents. Here's a simple example to get you started:

In following example, we import the BeautifulSoup class from the bs4 module. We then define a string variable that contains an HTML document, which we will parse using Beautiful Soup.

To create a BeautifulSoup object, we pass the HTML document and a parser to the BeautifulSoup constructor. In this case, we are using the html.parser, which is a built-in parser in Python.

 We then use the prettify() method to print the parsed HTML in a pretty format. This method indents the HTML to make it more readable.

````python
from bs4 import BeautifulSoup as bs

html_doc = """
<html>
<head>
  <title>My Web Page</title>
</head>
<body>
  <h1>Welcome to my Web Page</h1>
  <p>This is some text.</p>
  <ul>
    <li>Item 1</li>
    <li>Item 2</li>
    <li>Item 3</li>
  </ul>
</body>
</html>
"""

soup = bs(html_doc, 'html.parser') #HTML format -> data retrival is easy

print(soup.prettify())
````

As you can see, the parsed HTML is now indented and easier to read.

In addition to searching for tags, Beautiful Soup also provides various methods for searching for tags based on their attributes. Here's an example:

````python
html_doc = """
<html>
<head>
  <title>My Web Page</title>
</head>
<body>
  <h1>Welcome to my Web Page</h1>
  <p class="myclassname">This is some text.</p>
  <ul>
    <li class="baseclassname">Roti</li>
    <li>Chicken</li>
    <li>Paneer</li>
  </ul>
</body>
</html>
"""

soup = bs(html_doc,'html.parser')
p_tag = soup.find('p',class_='myclassname')
print(p_tag.text)

li_tag = soup.find_all('li')

for i in li_tag:
  print(i.text)
````

````python
html_doc = """
<html>
<head>
  <title>My Web Page</title>
</head>
<body>
  <h1>Welcome to my Web Page</h1>
  <p class="myclassname">This is some text.</p>
  <ul>
    <li class="baseclassname">Roti</li>
    <li>Chicken</li>
    <li>Paneer</li>
  </ul>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

p_tag = soup.find('p', class_='myclassname')
li_tags = soup.find_all('li', class_='baseclassname')

print(p_tag.text)

for li in li_tags:
    print(li.text)
````

## **Introduction to Requests**

üåêüì•üêç The Requests library is a powerful Python library used for making HTTP requests and handling responses from web servers. It's a popular choice for web scraping and can be used to retrieve web page content for further parsing and analysis. Here are some important things to know about using Requests for web scraping:

1Ô∏è‚É£ To start using Requests, you first need to install it. You can do this using pip, the package installer for Python. Open your command prompt and run the following command:

````python
!pip install requests
````

2Ô∏è‚É£ After installing Requests, you can start using it in your Python code. To make a GET request to a web page, you can use the get() method and pass in the URL of the page you want to retrieve:

````python
import requests
url = 'https://www.example.com'
response = requests.get(url)
````

3Ô∏è‚É£ The get() method returns a Response object, which contains the server's response to the request. You can access the content of the response using the text attribute, which returns the HTML content of the page:

````python
html_content = response.text
html_content
````

4Ô∏è‚É£ You can also pass additional parameters to the get() method to customize your request, such as headers or query parameters. For example, to pass query parameters in the URL:

````python
params = {'keywords': 'Data Science', 'txtLocation': 'value2'}
response = requests.get(url, params=params)
c=response.text
c
````

5Ô∏è‚É£ Requests also provides a convenient way to handle common authentication methods, such as Basic Authentication or OAuth. For example, to use Basic Authentication:

````python
auth = ('username', 'password')

response = requests.get(url, auth=auth)
r =response.text
r
````

6Ô∏è‚É£ Finally, it's important to handle errors that may occur when making HTTP requests. Requests raises exceptions for common HTTP errors, such as 404 Not Found or 500 Internal Server Error. You can handle these exceptions using a try...except block:

````python
try:
    response = requests.get(url)
    response.raise_for_status()
except requests.exceptions.HTTPError as err:
    print(f'HTTP error occurred: {err}')
except Exception as err:
    print(f'Other error occurred: {err}')
````

In summary, the Requests library is a powerful tool for making HTTP requests and retrieving web page content. It provides a simple and easy-to-use interface for handling common HTTP request and response features, such as headers, query parameters, and authentication. When using Requests for web scraping, it's important to handle errors that may occur and to follow ethical and legal guidelines for web scraping.

**Problem:**

Write a Python program to scrape the text from the Wikipedia page for "Python programming language" and display it on the console using Beautiful Soup.

###**Solution:**

Here's a Python program that uses Beautiful Soup to scrape the text from the Wikipedia page for "Python programming language" and display it on the console:

````python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
response = requests.get(url)

soup = BeautifulSoup(response.content,'lxml')

paragraphs  = soup.find_all('p')
for text in paragraphs:
  print(text.get_text().strip())
  print("**********")
````

This program first sends a GET request to the Wikipedia page for Python programming language and gets the HTML response. It then uses Beautiful Soup to parse the HTML response and find all the paragraphs on the page. For each paragraph, it extracts the text and displays it on the console.
